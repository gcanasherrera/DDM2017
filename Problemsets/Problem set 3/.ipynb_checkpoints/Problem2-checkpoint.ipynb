```python
def plot_histogram(data, binning_style, label, ax=None, xmin=10, xmax=17):
    """
    Convenience routine for plotting a histogram.
    """

    if ax is not None:
        ax = plt.axes(ax)

    counts, bins, patches = hist(data, bins=binning_style, ax=ax,
                                 color='k', histtype='step', normed=True)
    ax.text(0.95, 0.93, '{0:s}:\n{1:d} bins'.format(label, len(counts)),
            transform=ax.transAxes, ha='right', va='top')

    # Turn off y-axis labeling.
    ax.yaxis.set_major_formatter(plt.NullFormatter())

    ax.set_xlim(xmin, xmax)

    return ax
```

```python
np.random.seed(1)
N = 150
i1 = np.int(0.95*N)
i2 = np.int(0.05*N)
X = np.concatenate((np.random.normal(1.4, 0.35, i1),
                    np.random.normal(2.55, 0.05, i2)))[:, np.newaxis]
                   
np.random.shuffle(X)

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111)
plot_histogram(X[:,0], 10, 'Gaussian Reconstruction', ax=ax, xmin=np.min(t['Mass'])-1, xmax=np.max(t['Mass'])+1)
#plot_histogram(t['Mass'], 10, 'Gaussian Reconstruction', ax=ax, xmin=np.min(t['Mass'])-1, xmax=np.max(t['Mass'])+1)
```

```python
from astroML.density_estimation import KNeighborsDensity
from sklearn.cross_validation import KFold
from sklearn.neighbors import KernelDensity
from astroML.datasets import fetch_great_wall
from matplotlib.colors import LogNorm

def cv1(x, bws, model='gaussian', plot=False, n_folds=10):
    """
    This calculates the leave-one-out cross validation. If you set 
    plot to True, then it will show a big grid of the test and training
    samples with the KDE chosen at each step. You might need to modify the 
    code if you want a nicer layout :)
    """

    # Get the number of bandwidths to check and the number of objects
    N_bw = len(bws)
    N = len(x)
    cv_1 = np.zeros(N_bw)
    
    # If plotting is requested, set up the plot region
    if plot:
        fig, axes = plt.subplots(N_bw, np.ceil(N/n_folds), figsize=(15, 8))
        xplot = np.linspace(-3, 8, 1000)

    # Loop over each band-width and calculate the probability of the 
    # test set for this band-width
    for i, bw in enumerate(bws):
    
        # I will do N-fold CV here. This divides X into N_folds
        kf = KFold(N, n_folds=n_folds)

        # Initiate - lnP will contain the log likelihood of the test sets
        # and i_k is a counter for the folds that is used for plotting and
        # nothing else..
        lnP = 0.0
        i_k = 0
                                 
        # Loop over each fold
        for train, test in kf:
            x_train = x[train, :]
            x_test = x[test, :]
            
            # Create the kernel density model for this bandwidth and fit
            # to the training set.
            kde = KernelDensity(kernel=model, bandwidth=bw).fit(x_train)
                                 
            # score evaluates the log likelihood of a dataset given the fitted KDE.
            log_prob = kde.score(x_test)
            
            if plot:
                # Show the tries
                ax = axes[i][i_k]

                # Note that the test sample is hard to see here.
                hist(x_train, bins=10, ax=ax, color='red')
                hist(x_test, bins=10, ax=ax, color='blue')
                ax.plot(xplot, np.exp(kde.score_samples(xplot[:, np.newaxis])))
                i_k += 1
            

            lnP += log_prob
            
        # Calculate the average likelihood          
        cv_1[i] = lnP/N
        
    return cv_1
```

```python
bws = np.linspace(0.001, 11, 5000)
cv = cv1(X, bws)
#print np.exp(cv)
plt.plot(bws, np.exp(cv))
plt.xlabel('bandwidth')
plt.ylabel('CV likelihood')
plt.text(5, 0.0004, 'Best BW={0:.4f}'.format(bws[np.argmax(cv)]))

#La conclusion: este método hace overfitting y básicamente estamos j**
```

```python
#Reconstruction kernel
from astroML.density_estimation import KDE

kde = KDE('gaussian', 0.1)
x = t['Mass']
X = x[:, np.newaxis]
kde.fit(X)
Xgrid = np.linspace(0, 3, 100)[:, np.newaxis]
dens  = kde.eval(Xgrid)
plt.plot(Xgrid[0:], dens)
```

```python
dens_norm = dens/sum(dens)
print sum(dens_norm)
plt.plot(Xgrid[0:], dens_norm)
```

```python
#Let's integrate to calculate the finding a neutron star larger than 1.8 in %
from scipy.integrate import simps
simps(dens_norm[np.where(Xgrid>1.8)[0]])*100
```

```python
#(b) As the KDE are gaussian, the probability can be summed up
mass1=simps(dens_norm[np.where(np.logical_and(Xgrid>1.36, Xgrid<2.26))[0]])
mass2=simps(dens_norm[np.where(np.logical_and(Xgrid>0.86, Xgrid<1.36))[0]])

print mass1
print mass2

binary = mass1+mass2
print 'sum={0}'.format(binary)
```

```python
# (c) We need to simulate 5 random points of our distribution, but our distribution is with astroML and not with Kernel
#Density.... Repear for KernelDensity package

from sklearn.neighbors import KernelDensity
kernel = KernelDensity(bandwidth=0.05, kernel='gaussian')
x_kd = t['Mass']
X_kd = x_kd[:, np.newaxis]
kernel.fit(X)
Xgrid_kd = np.linspace(0.01, 3, 100)[:, np.newaxis]
dens_kd  = kernel.score_samples(Xgrid_kd)
dens_kd = np.exp(dens_kd)
plt.plot(Xgrid_kd[0:], dens_kd)
```
